```python
# 1. Convert to datetime
df['release_date'] = pd.to_datetime(df['release_date'], dayfirst=True)

# 2. Extract numeric features
df['rel_month'] = df['release_date'].dt.month
df['rel_day'] = df['release_date'].dt.day

# 3. Drop the original column (Models can't use it)
df.drop('release_date', axis=1, inplace=True)
```

```python
# 1. Split the strings by '|' and explode them into individual rows
genre_counts = df['genre'].str.split('|').explode().value_counts()

# 2. View the distribution
print(genre_counts)

# 3. See the percentage (to decide what to cut)
print(genre_counts / len(df) * 100)
```

```python
# 1. Define your 'Major' genres based on your data (Top 10)
major_genres = ['Drama', 'Thriller', 'Action', 'Comedy', 'Romantic', 
                'Crime', 'Adventure', 'Horror', 'Family', 'Mystery']

# 2. Function to clean and group
def simplify_genres(genre_string):
    if pd.isna(genre_string):
        return ['Other_Genre']

    genres = genre_string.split('|')
    refined = [g if g in major_genres else 'Other_Genre' for g in genres]

    return list(set(refined))


# 3. Apply and Binarize
df['genre_list'] = df['genre'].apply(simplify_genres)

from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(df['genre_list'])

# This creates only 11 columns (10 major + 1 'Other') instead of 35!
genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_, index=df.index)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd

# 1. Initialize and Fit the Binarizer on your existing 'genre_list'
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(df['genre_list'])

# 2. Create the binary DataFrame
genre_columns = pd.DataFrame(genre_encoded, columns=mlb.classes_, index=df.index)

# 3. Concatenate and Drop the text-heavy columns
# We drop 'genre' (the string) and 'genre_list' (the list object)
df = pd.concat([df, genre_columns], axis=1).drop(['genre', 'genre_list'], axis=1)

# View the result
df.head()
```

```python
# 1. Split the strings by '|' and explode them into individual rows
languages_counts = df['languages'].str.split('|').explode().value_counts()

# 2. View the distribution
print(languages_counts)

# 3. See the percentage (to decide what to cut)
print(languages_counts / len(df) * 100)
```

```python
# 1. Define your 'Major' languages based on your data
major_languages = ['Telugu', 'Tamil', 'Hindi', 'Kannada', 'Malayalam','English']

# 2. Function to clean and group
def simplify_languages(language_string):
    if pd.isna(language_string):
        return ['Other_Language']

    languages = language_string.split('|')
    refined = [g if g in major_languages else 'Other_Language' for g in languages]

    return list(set(refined))


# 3. Apply and Binarize
df['language_list'] = df['languages'].apply(simplify_languages)

from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
language_encoded = mlb.fit_transform(df['language_list'])
# This creates only 11 columns (10 major + 1 'Other') instead of 35!
language_df = pd.DataFrame(language_encoded, columns=mlb.classes_, index=df.index)
```
